```python
import symtable as syt
```

```python
def get_table(filename):
    with open(filename, 'r') as file:
        code = file.read()
        st = syt.symtable(code, filename=filename, compile_type='exec')
    return st
```

```python
st = get_table('code_example.py')
```

```python
def rec(root, dpt=0):
    print('\t'*dpt, root, type(root))
    for mth in [x for x in dir(root) if x.startswith('is') or x.startswith('get')]:
        print('\t'*(dpt+1), mth, ':', getattr(root, mth)())

    for symbol in root.get_symbols():
        print('\t'*(dpt+1), symbol)
        for mth in [x for x in dir(symbol) if x.startswith('is') or x.startswith('get')]:
            try:
                print('\t'*(dpt+2), mth, ':', getattr(symbol, mth)())
            except ValueError:
                pass
    
    for table in root.get_children():
       rec(table, dpt+1) 
```

```python
rec(st)
```
